```python
import os
```

```python
trainHorseDir = os.path.join('./dataSets/horseVsHuman/horse-or-human/horses/')
```

```python
trainHumanDir = os.path.join('./dataSets/horseVsHuman/horse-or-human/humans/')
```

```python
trainHorseNames = os.listdir(trainHorseDir)
print(trainHorseNames[:10])
```

```python
trainHumanNames = os.listdir(trainHumanDir)
print(trainHumanNames[:10])
```

```python
print('total training horse images:', len(os.listdir(trainHorseDir)))
print('total training human images:', len(os.listdir(trainHumanDir)))
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
numberOfRows = 4
numberOfColumns = 4
```

```python
picIndex = 0
```

```python
fig = plt.gcf()
```

```python
fig.set_size_inches(numberOfColumns*4, numberOfRows*4)
```

```python
picIndex += 8
nextHorsePix = [os.path.join(trainHorseDir, fname) for fname in trainHorseNames[picIndex - 8:picIndex]]
nextHumanPix = [os.path.join(trainHumanDir, fname) for fname in trainHumanNames[picIndex - 8:picIndex]]
for i, img_path in enumerate(nextHorsePix+nextHumanPix):
  # Set up subplot; subplot indices start at 1
  sp = plt.subplot(numberOfRows, numberOfColumns, i + 1)
  sp.axis('Off') # Don't show axes (or gridlines)
  img = mpimg.imread(img_path)
  plt.imshow(img)
plt.show()
```
